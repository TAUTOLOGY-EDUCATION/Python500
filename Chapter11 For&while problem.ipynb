# - Question 11

### for loop

```python
n = int(input('Please insert number : '))
sumx = 0
for i in range(1, n+1):
    sumx = sumx + i
print(sumx)
```

### while loop

```python
n = int(input('Please insert number : '))
sumx = 0
i = 1
while i <= n:
    sumx = sumx + i
    i = i + 1
print(sumx)
```

# - Question 15

### for loop

```python
n = int(input('Please insert number : '))
sumx = 0
for i in range(1, n+1):
    sumx = sumx + 1/i**2
pi = (6*sumx)**(1/2)
print(pi)
```

### while loop

```python
n = int(input('Please insert number : '))
sumx = 0
i = 1
while i <= n:
    sumx = sumx + 1/i**2
    i = i + 1
pi = (6*sumx)**(1/2)
print(pi)
```
